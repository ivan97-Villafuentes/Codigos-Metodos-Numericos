## Esquema Numéricos de Primer Orden en Tiempo

Aproximemos el siguiente esquema numérico Semi-implícito

 $\frac{u_{n}-u_{n-1}}{\Delta t}+\theta u_{n} + (1-\theta)u_{n-1}=0$ con $\theta \in [0,1]$

```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
dt = 1.2
T = 100
k=0
N = int(T/dt)
t = np.zeros(N+1)
u = np.zeros(N+1)
print(N)


U0 = 5         #Condicion inicial
t[0] = 0       
u[0] = U0

def exact(t):                # Definimos  nuestra solucion exacta asociada a nuestro problema
    return U0*np.exp(-t)

for n in range(N):
    t[n+1] = t[n] + dt                        #Paso del tiempo
    u[n+1] = ((1 - dt*(1-k))/(1+dt*k))*u[n]    #Esquema Numerico
  
plt.plot(t,u, label='Aprox Numerica')
plt.plot(t,exact(t), label='Sol Exacta')    
plt.title("Aproximacion Solucion metodo Semi-implicito")
plt.legend()
plt.grid()
plt.show()
```


##  Estimación de la energía

```python

print(t)
EE= np.zeros(N+1) #Vector
EES=[]
for i in range(N):
    t[i+1] = t[i] + dt
    EE=((u[i+1]-u[i]/dt)**2)+(1/2*(u[i+1]**2-u[i]**2/dt))+((dt/2)*(2*k-1)*(u[i+1]-u[i]/dt)**2) ##  la ley de energia 
    EE1=EE
    EES.append(EE1)
plt.plot(EES,label='Energia')          #Grafica de Energia
plt.title("Estabilidad Energetica")
plt.legend()
plt.grid()
plt.show()
    
```
